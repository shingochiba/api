## 特徴

- このバージョンは 初心者〜中級者向けにフル機能の AUTOMATIC1111 Stable Diffusion WebUI が使えるようになっています。
- お試しとして一次領域を使うことで、ストレージ(Google Drive)の消費を抑える設定になっています。
- より初心者向けには選択肢が少ない [Fooocus日本語版](https://j.aicu.ai/FoooC) をお勧めします。
- 生成した画像は「マイドライブ/sd/stable-diffusion-webui/outputs」フォルダに保存されます。
- 設定や流れは書籍やnoteの記事に沿っています。このスクリプトをベースに、よりシンプルにしたり、順番を変えたりといった工夫はご自身で複製してやってみてください。

## 使い方

1. [Google Colab Pro (有料)](https://aicu.page.link/ColabPrice)で T4環境以上のGPUを利用できるGoogleアカウントでご利用ください（それ以外の環境でも動作はする可能性がありますが、途中で停止するなどの不具合が発生する可能性があります）。
1. 文末で紹介しているURLに行き、ボタン「Open in Colab」を押します。
1. ファイル→「ドライブにコピーを保存」でお好きな名前で保存してください。
1. 以下に続くコードセル「**Connect Google Drive**」左にある「▶」を上から順に押していってください。
1. Googleドライブへの接続を確認されますので、上記のGoogleアカウントで接続してください。ドライブは最低でも数GB必要になります。
1. 「Start Stable Diffusion」のセルを実行して表示される「https://xxxxxxxxx.gradio.live/ 」のURLにアクセスすることで **AUTOMATIC1111/Stable Diffusion WebUI** が利用できます。
1. 生成した画像は「マイドライブ/sd/stable-diffusion-webui/outputs」フォルダに保存されます。
1. 作業を終了する場合は「■」ボタンで止めて、このファイル（Google Colab）を閉じて構いません。
